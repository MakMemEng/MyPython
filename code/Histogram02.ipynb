```python
import cv2
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Joker.jpg')
```

```python
color = ('b','g','r')
```

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
```

```python
plt.show()
```
