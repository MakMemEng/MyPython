```python
import pandas as pd
df = pd.read_csv('./input/data1.csv')

# 全データの表示
df
```

```python
# 欠損値(NaN) を含む"行"を削除
df1 = df.dropna()
df1
```

```python
# 欠損値(NaN) を含む"列"を削除
df2 = df.dropna(axis=1)
df2
```

```python
# 欠損値(NaN)ではない値が4つ未満の"行"を削除
df3 = df.dropna(thresh=4)  # 亀子が消える
df3
```

```python
# 特定の列(身長)に欠損値(NaN)を含む"行"を削除
df4 = df.dropna(subset=["身長(cm)"])
df4
```
